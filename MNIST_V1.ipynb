# PreProcessing Data

## Labels

* We first need to make sure the labels will be understandable by our CNN.
* Right now our labels are just literally the number they represent.
* If we were to pass in our training labels as such, the network will assume it was a some sort of continuous value, and it would try to predict stuff like 5.5 or 5.6
* But really these are categories. So 5is not a continuous value but a category 5.
* So really what we are doing here is a classification problem.
* We need to make sure our network understands that.
* We need to OneHotEncode this.
* Luckily TensorFlow has utilities for this. (to_categorical).

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train.shape
```

```python
y_cat_train = to_categorical(y_train, num_classes= 10)
```

```python
y_cat_train.shape
```

* It has now automatically with this to_categorical function call OneHotEncoded this.

```python
y_cat_train[0]
```

* See what happens at index position 5 its now 1,essentially indicating that this belongs to class 5.

```python
y_cat_test = to_categorical(y_test, num_classes= 10)
```

### Processing X Data

```python
X_train.max()
```

```python
X_train.min()
```

* If we check the max and min value it is 255 and 0 respectively.
* To make sure we don't run into gradient problems these should be scaled between 0 & 1.

```python
X_train = X_train/255
X_test = X_test/255
```

```python
# display or plot a number.
plt.figure(figsize=(7,7))
idx = 61

grid_data = X_train[idx]  # reshape from 1d to 2d pixel array
plt.imshow(grid_data, interpolation = "none", cmap = "gray_r")
plt.show()

print(y_train[idx])
```

## Reshaping the Data

```python
X_train.shape
```

```python
X_test.shape
```

* Right now our data is 60,000 images stored in 28 by 28 pixel array formation. 

* This is correct for a CNN, but we need to add one more dimension to show we're dealing with single RGB channel (since technically the images are in black and white, only showing values from 0-255 on a single channel), an color image would have 3 dimensions.

```python
# batch_sze,width,height,color_channels
X_train = X_train.reshape(60000, 28, 28, 1)
X_train.shape
```

```python
X_test = X_test.reshape(10000,28,28,1)
X_test.shape
```

# Building the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model= Sequential()

# CONVOLUTIONAL LAYER
model.add(Conv2D(filters= 32, kernel_size= (4,4), input_shape= (28, 28, 1), activation= 'relu'))

# POOLING LAYER
model.add(MaxPool2D(pool_size= (2, 2)))

# FLATTEN IMAGES FROM 28 by 28 to 784 BEFORE FINAL LAYER
model.add(Flatten())

# 128 NEURONS IN DENSE HIDDEN LAYER (YOU CAN CHANGE THIS NUMBER OF NEURONS)
model.add(Dense(128, activation= 'relu'))

# LAST LAYER IS THE CLASSIFIER, THUS 10 POSSIBLE CLASSES
model.add(Dense(10, activation= 'softmax'))
```

```python
# https://keras.io/metrics/

model.compile(loss= 'categorical_crossentropy', optimizer= 'adam',\
              metrics= ['accuracy']) # we can add in additional metrics https://keras.io/metrics/
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop= EarlyStopping(monitor= 'val_loss', patience= 2)
```

## Train the Model

```python
model.fit(x= X_train,
          y= y_cat_train,
          epochs= 10,
          validation_data= (X_test, y_cat_test),
          callbacks= [early_stop]
         )
```

## Evaluating the Model

```python
model.metrics_names
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.head()
```

```python
plt.style.use("fivethirtyeight")
losses[['accuracy','val_accuracy']].plot(figsize= (10,7))
```

```python
losses[['loss','val_loss']].plot(figsize= (10,7))
```
